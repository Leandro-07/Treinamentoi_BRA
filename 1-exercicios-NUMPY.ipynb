### 1. Criar um array com 10 números inteiros aleatórios entre 1 e 50 e calcular a média e o desvio padrão

```
import numpy as np

# Criar um array com 10 numeros inteiros aleatorios entre 1 e 50
arr = np.random.randint(1,51, size=10)

# Calcular a media e o desvio padrão do array
media = np.mean(arr)
desvio_padrao = np.std(arr)

print("Array: ", arr)
print("Média: ", media)
print("Desvio padrão: ", desvio_padrao)
```

### 2. Criar um array de 10 números aleatórios e calcular a raiz quadrada de todos os valores menores ou iguais a 0,5

```
arr = np.random.rand(10)

# Calcular a raiz quadrada dos valores menores ou iguais a 0,5
arr[arr <= 0.5] = np.sqrt(arr[arr <= 0.5])

print(arr)
```

### 3. Criar uma matriz 3D de tamanho 3x3x3 com valores aleatórios e exibir o valor máximo de cada linha de cada matriz 2D dentro dessa matriz 3D

```
matriz_3D = np.random.rand(3, 3, 3)

#Exibir o valor maxima de cada linha de cada matriz 2D dentro da matriz 3D
max_valores = np.amax(matriz_3D, axis=1)


print("Matriz 3D: \n", matriz_3D)
print("Valores maiores de cada linha da matriz 3D: \n", max_valores)
```

### 4. Criar uma matriz de tamanho 5x5 com valores aleatórios e subtrair a média de cada linha dessa matriz

```
matriz = np.random.rand(5,5)

media = np.mean(matriz, axis=1)
media_linha = matriz - media[:, np.newaxis]

print("Matriz gerada: \n", matriz)
print("Média da linha: \n", media)
print("Matriz com a media das linhas subtraidas: \n", media_linha)
```

### 5. Criar uma matriz de tamanho 4x4 com valores aleatórios e calcular a média e o desvio padrão das diagonais da matriz

```
import numpy as np
matriz = np.random.rand(4,4)

diagonal_principal = np.diag(matriz)
diagonal_secundaria = np.diag(np.fliplr(matriz))
media_diagonal = np.mean([diagonal_principal, diagonal_secundaria])
std_diagonal = np.std([diagonal_principal, diagonal_secundaria])

print("Matriz original: \n", matriz)
print("Diagonal Princial: \n", diagonal_principal)
print("Diagonal Secundária: \n", diagonal_secundaria)
print("Media da Diagonal: \n", media_diagonal)
print("Desvio Padrão: \n", std_diagonal)
```
